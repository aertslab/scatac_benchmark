## Export DARs

```python
f_dar_list = {
    'fulldata': '/staging/leuven/stg_00002/lcb/cflerin/analysis/pbmc_atac/analysis2/data_freeze_Jun2021/atac_qc_multiplet_merged/jupyter/downstream_analysis/markers_dict.pkl',
    'cellds': '/staging/leuven/stg_00002/lcb/cflerin/analysis/pbmc_atac/analysis2/data_freeze_Jun2021_cell_downsampled/atac_qc/jupyter/downstream_analysis/markers_dict.pkl',
    'libds': '/staging/leuven/stg_00002/lcb/cflerin/analysis/pbmc_atac/analysis2/data_freeze_Jun2021_library_downsampled/atac_qc_multiplet_merged/jupyter/downstream_analysis/markers_dict.pkl'
}

f_dar_output_list = {
    'fulldata': 'rerun_cflerin_dar_script_fulldata/',
    'cellds': 'rerun_cflerin_dar_script_cellds/',
    'libds': 'rerun_cflerin_dar_script_libds/'
}


mdict = {}
for ds in f_dar_list.keys():
    with open(f_dar_list[ds], 'rb') as f:
        mdict[ds] = pickle.load(f)
```

```python
for ds in mdict.keys():
    print(ds)
    for key in mdict[ds].keys():
        print(' '+key)
        for ct in mdict[ds][key].keys():
            #print('  '+ct)
            s = 'merged' if key=='merged' else sample_map[key]
            f_out = os.path.join(f_dar_output_list[ds], ds + '__' + s + '__' + re.sub(" ", "_", ct) + '__DARs.txt')
            pd.concat([
                pd.DataFrame(
                    [ re.split(':|-',x) for x in mdict[ds][key][ct].index ],
                    columns=['chrom','chromStart','chromEnd'],
                    index=mdict[ds][key][ct].index,
                ),
                mdict[ds][key][ct]
            ],
                axis=1
            ).to_csv(f_out, sep='\t', index=False, header=False)
```
