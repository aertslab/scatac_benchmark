### Read in LDA/mallett models

```python
from pycisTopic.lda_models import evaluate_models
import matplotlib.pyplot as plt
```

```python
f_mod_dir = 'pycistopic_consensus_peaks/models__consensus__mallet/'
n_iter = 500
models_dict = {}
for key in ['libds_merged']:
    f_mod =os.path.join(wdir, f_mod_dir, key + '__models_' + str(n_iter) + '_iter.pkl')
    if(os.path.isfile(f_mod)):
        with open(f_mod, 'rb') as f:
            models_dict[key] = pickle.load(f)
        print(f"Loaded {key}")
    else:
        print(f"file {f_mod} doesn't exist")
```

```python
for key in models_dict.keys():
    model=evaluate_models(models_dict[key],
                         select_model=90, 
                         return_model=True, 
                         metrics=['Arun_2010','Cao_Juan_2009', 'Minmo_2011', 'loglikelihood'],
                         plot=True,
                         plot_metrics=False
                         )
    cistopic_obj_dict[key].add_LDA_model(model)
```

```python
lisi['pre-harmony_umap']['group'] = 'Pre-harmony UMAP'
lisi['pre-harmony_umap']['sample_id'] = cistopic_obj_dict['libds_merged'].cell_data['sample_id']
lisi['pre-harmony_umap']['alias'] = cistopic_obj_dict['libds_merged'].cell_data['alias']

lisi['post-harmony_umap']['group'] = 'Post-harmony UMAP'
lisi['post-harmony_umap']['sample_id'] = cistopic_obj_dict['libds_merged'].cell_data['sample_id']
lisi['post-harmony_umap']['alias'] = cistopic_obj_dict['libds_merged'].cell_data['alias']

lisi['pre-harmony_tsne']['group'] = 'Pre-harmony tSNE'
lisi['pre-harmony_tsne']['sample_id'] = cistopic_obj_dict['libds_merged'].cell_data['sample_id']
lisi['pre-harmony_tsne']['alias'] = cistopic_obj_dict['libds_merged'].cell_data['alias']

lisi['post-harmony_tsne']['group'] = 'Post-harmony tSNE'
lisi['post-harmony_tsne']['sample_id'] = cistopic_obj_dict['libds_merged'].cell_data['sample_id']
lisi['post-harmony_tsne']['alias'] = cistopic_obj_dict['libds_merged'].cell_data['alias']
```

```python
def cilo(x):
    return x.mean() - 1.95 * x.sd() / math.sqrt(x.shape[0])
```

```python
ldf = pd.concat([
    lisi['pre-harmony_umap'].groupby('sample_id').mean(),
    lisi['post-harmony_umap'].groupby('sample_id').mean(),
    lisi['pre-harmony_tsne'].groupby('sample_id').mean(),
    lisi['post-harmony_tsne'].groupby('sample_id').mean(),
],axis=1)
ldf.columns = ['Pre-harmony UMAP', 'Post-harmony UMAP', 'Pre-harmony tSNE', 'Post-harmony tSNE']
```

```python
ldf['Pre-harmony UMAP CI'] = ''
ldf['Post-harmony UMAP CI'] = ''
ldf['Pre-harmony tSNE CI'] = ''
ldf['Post-harmony tSNE CI'] = ''
```

```python
def get_formatted_ci(x, i):
    ix = x['sample_id'] == i
    s = x[ix]['lisi'].std()
    m = x[ix]['lisi'].mean()
    
    return (#str(round(m,3)) + 
            ' (' +
        str(round(m - 1.96 * s / math.sqrt(len(ix)),3)) +
        ' - ' +
        str(round(m + 1.96 * s / math.sqrt(len(ix)),3)) +
        ')')

for i in ldf.index:
    ldf.loc[i,'Pre-harmony UMAP CI'] = get_formatted_ci(lisi['pre-harmony_umap'], i)
    ldf.loc[i,'Post-harmony UMAP CI'] = get_formatted_ci(lisi['post-harmony_umap'], i)
    ldf.loc[i,'Pre-harmony tSNE CI'] = get_formatted_ci(lisi['pre-harmony_tsne'], i)
    ldf.loc[i,'Post-harmony tSNE CI'] = get_formatted_ci(lisi['post-harmony_tsne'], i)
```

```python
# reorder legends function
import numpy as np
#  Returns tuple of handles, labels for axis ax, after reordering them to conform to the label order `order`, and if unique is True, after removing entries with duplicate labels.
def reorderLegend(ax=None,order=None,unique=False):
    if ax is None: ax=plt.gca()
    handles, labels = ax.get_legend_handles_labels()
    labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0])) # sort both labels and handles by labels
    if order is not None: # Sort according to a given list (not necessarily complete)
        keys=dict(zip(order,range(len(order))))
        labels, handles = zip(*sorted(zip(labels, handles), key=lambda t,keys=keys: keys.get(t[0],np.inf)))
    if unique:  labels, handles= zip(*unique_everseen(zip(labels,handles), key = labels)) # Keep only the first of each handle
    ax.legend(handles, labels, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # i modified this line 

    return(handles, labels)


def unique_everseen(seq, key=None):
    seen = set()
    seen_add = seen.add
    return [x for x,k in zip(seq,key) if not (k in seen or seen_add(k))]


#source: https://stackoverflow.com/questions/22263807/how-is-order-of-items-in-matplotlib-legend-determined
```

```python
tp = pd.concat([
    lisi['pre-harmony_topic'],
    lisi['post-harmony_topic'],
    lisi['pre-harmony_umap'],
    lisi['post-harmony_umap'],
    lisi['pre-harmony_tsne'],
    lisi['post-harmony_tsne'],
])

fig, ax = plt.subplots(1,1, figsize=(10,5), dpi=150)

g = sns.boxplot(data=tp, x='group', y='lisi',
                   hue='tech',
                   palette=color_dict_tech,
                   dodge=True,
                   order=order,
                   #linewidth=0.1,
                   fliersize=1,
                   #inner='quartiles',
                   ax=ax,# kind='kde',
                   #width=0.9
                   )

reorderLegend(ax,reorder)

ax.set_xlabel('')
ax.set_ylabel('LISI')
ax.legend(loc=2)
g.set_xticklabels(g.get_xticklabels(), rotation=15, ha='right', fontsize=10)
plt.tight_layout()
plt.savefig('plts_pub/lisi_allspaces_bytech_grouped.png', dpi=300, facecolor='white')
plt.savefig('plts_pub/lisi_allspaces_bytech_grouped.svg', dpi=300, facecolor='white')
```
